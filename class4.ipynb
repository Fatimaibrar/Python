```python
#Create a function that accepts a list of numbers and returns the largest number from the list.
#Example:
def largest_number(numbers):
    return max(numbers)

# Example
nums = [10, 45, 67, 23]
print("Largest number:", largest_number(nums))
```

```python
#Write a function that takes a string as input and counts the number of vowels in it.
# Example:
# "Python Programming"

def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    
    for char in text:
        if char in vowels:
            count += 1
            
    return count

# Example
string = "Python Programming"
print("Number of vowels:", count_vowels(string))
```

```python
#Create a lambda function that checks whether a number is:
# •	Even 
# •	Odd 
# Using if-else inside lambda
is_even = lambda x: True if x % 2 == 0 else False
is_odd = lambda x: True if x % 2 != 0 else False
# Example
number = 10
print("Is even:", is_even(number))
print("Is odd:", is_odd(number))
```

```python
account = {
    "name": "Fatima",      
    "balance": 7000     
}

# Deposit function
def deposit(amount):
    if amount > 0:                          
        account["balance"] += amount         
        print("Deposited:", amount)
        print("New Balance:", account["balance"])
    else:
        print("Invalid amount!")             

# Withdraw function
def withdraw(amount):
    if amount > account["balance"]:           
        print("Insufficient balance!")        
    elif amount <= 0:                         
        print("Invalid amount!")
    else:
        account["balance"] -= amount          
        print("Withdrawn:", amount)
        print("Remaining Balance:", account["balance"])

# Balance check function
def check_balance():
    print("Account Holder:", account["name"])  
    print("Current Balance:", account["balance"]) 

# Banking system menu loop
def banking_menu():
    while True:                         
        print("\n--- Banking Menu ---")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Exit")

        choice = input("Enter choice: ") 
        if choice == "1":
            amt = int(input("Enter deposit amount: "))
            deposit(amt)                                
        elif choice == "2":
            amt = int(input("Enter withdraw amount: ")) 
            withdraw(amt)                               
        elif choice == "3":
            check_balance()                            
        elif choice == "4":
            print("Thank you! Goodbye.")
            break
        else:
            print("Wrong choice! Try again.")  

print("Welcome to the Banking System!")
print("Account Holder:", account["name"])
banking_menu()
```

```python
#Product Management System
products = {}   

# Product add karo
def add_product(name, price):
    products[name] = price              
    print(name, "added successfully!")

# Product remove karo
def remove_product(name):
    if name in products:              
        del products[name]             
        print(name, "removed!")
    else:
        print(name, "not found!")      
# Saare products display karo
def display_products():
    if len(products) == 0:            
        print("No products available.")
    else:
        print("\n--- Product List ---")
        for name, price in products.items():  
            print(f"{name} : Rs. {price}")    

def total_price():
    total = 0                         
    for price in products.values():    
        total += price                 
    print("Total Price of all products: Rs.", total)

# Product system demo
add_product("Milk", 210)
add_product("Bread", 150)
add_product("Eggs", 350)
display_products()
total_price()
remove_product("Eggs")
display_products()
```
