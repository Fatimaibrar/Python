# Operators

```python
# take two numbers from the user and perform all arithmetic operations.
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
if num2 != 0:
    print("Division:", num1 / num2)
print("Modulus:", num1 % num2)
print("Exponentiation:", num1 ** num2)
```

```python
#Compare two numbers using comparison operators.
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
print("Equal:", num1 == num2)
print("Not Equal:", num1 != num2)
print("Greater Than:", num1 > num2)
print("Less Than:", num1 < num2)
print("Greater Than or Equal:", num1 >= num2)
print("Less Than or Equal:", num1 <= num2)
```

# Loops

```python
#. Use nested loops to print a rectangle star pattern.
rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

for i in range(rows):
    for j in range(cols):
        print("*", end=" ")
    print()
   
```

```python
#Print multiplication tables using nested loops.
for i in range(1, 11):
    for j in range(1, 11):
        print(i * j, end=" ")
    print()
```

```python
# . Print a triangle number pattern using nested loops.
rows = int(input("Enter number of rows: "))
for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
# Count vowels in a string using loops.
string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for char in string:
    if char in vowels:
        count += 1
print("Number of vowels in the string:", count)
```

```python
# Shopping Cart System

USERS = {"admin": "pass123", "ali": "abc123", "maria": "xyz789"}

PRODUCTS = {
    "Apple": 50, "Banana": 20, "Milk": 120,
    "Bread": 70, "Eggs": 10, "Chocolate": 150, "Juice": 90
}

cart = {}  # {product_name: quantity}

def login():
    print("=== Shopping Cart System ===")
    for _ in range(3):
        u = input("Username: ").strip()
        p = input("Password: ").strip()
        if USERS.get(u) == p:
            print(f"Welcome, {u}!\n")
            return u
        print("Wrong credentials. Try again.\n")
    print("Too many attempts. Exiting.")
    return None


def show_products():
    print("Products:")
    for name, price in PRODUCTS.items():
        print(f"  {name}: Rs.{price}")
    print()


def add_item():
    show_products()
    raw = input("Enter product and quantity (e.g. Apple,2,Milk,1): ").strip().split(",")
    raw = [x.strip() for x in raw]
    if len(raw) % 2 != 0:
        print("Invalid input.\n")
        return
    for i in range(0, len(raw), 2):
        name, qty = raw[i].title(), raw[i+1]
        if name not in PRODUCTS:
            print(f"'{name}' not found.")
            continue
        if not qty.isdigit() or int(qty) <= 0:
            print(f"Invalid quantity: {qty}")
            continue
        cart[name] = cart.get(name, 0) + int(qty)
        print(f"Added {qty} x {name}.")
    print()


def remove_item():
    name = input("Product to remove: ").strip().title()
    if name in cart:
        del cart[name]
        print(f"Removed {name}.\n")
    else:
        print(f"{name} not in cart.\n")


def update_item():
    name = input("Product to update: ").strip().title()
    if name not in cart:
        print("Not in cart.\n")
        return
    qty = input("New quantity: ").strip()
    if qty.isdigit() and int(qty) > 0:
        cart[name] = int(qty)
        print(f"Updated {name} to {qty}.\n")
    else:
        print("Invalid quantity.\n")


def view_cart():
    if not cart:
        print("Cart is empty.\n")
        return
    print("Your cart:")
    for name, qty in cart.items():
        print(f"  {name}: {qty} x Rs.{PRODUCTS[name]} = Rs.{qty * PRODUCTS[name]}")
    print()


def checkout(username):
    if not cart:
        print("Cart is empty.\n")
        return

    total = sum(PRODUCTS[n] * q for n, q in cart.items())
    discount = 0.10 if total >= 500 else 0.05 if total >= 300 else 0
    final = total * (1 - discount)
    status = "Gold" if total >= 500 else "Silver" if total >= 300 else "Regular"

    print(f"""
======== BILL ========
Customer : {username}
Items    : {', '.join(cart)}
Total    : Rs.{total}
Discount : {int(discount*100)}% = Rs.{total*discount:.2f}
To Pay   : Rs.{final:.2f}
Status   : {status} Customer
======================
""")


def menu():
    options = {
        "1": ("Show products", show_products),
        "2": ("Add to cart", add_item),
        "3": ("Remove from cart", remove_item),
        "4": ("Update quantity", update_item),
        "5": ("View cart", view_cart),
    }
    user = login()
    if not user:
        return
    while True:
        for key, (label, _) in options.items():
            print(f"{key}. {label}")
        print("6. Checkout\n7. Exit")
        choice = input("Choose (1-7): ").strip()
        if choice in options:
            options[choice][1]()
        elif choice == "6":
            checkout(user)
            break
        elif choice == "7":
            print("Goodbye!")
            break
        else:
            print("Invalid choice.\n")
menu()
```
