```python

# Login Management System 
users = []  

is_valid_password = lambda pwd: len(pwd) >= 6

def load_users():
    global users
    users = []

    try:
        with open("users.txt", "r") as file:

            for line in file:
                data = line.strip().split(",")

                if len(data) == 4:
                    username, password, age, email = data

                    users.append({
                        "username": username,
                        "password": password,
                        "age": int(age),
                        "email": email
                    })

    except FileNotFoundError:
        users = []
def save_users():

    with open("users.txt", "w") as file:

        for user in users:

            file.write(
                f"{user['username']},{user['password']},{user['age']},{user['email']}\n"
            )
def register():

    print("\n--- Register New Account ---")

    username = input("Enter username: ").strip()

    if not username:
        print("Error: Username cannot be empty!")
        return

    for user in users:
        if user["username"] == username:
            print("Error: Username already exists!")
            return

    password = input("Enter password (min 6 chars): ").strip()

    if not password:
        print("Error: Password cannot be empty!")
        return

    if not is_valid_password(password):
        print("Error: Password too short!")
        return

    try:
        age = int(input("Enter age: "))

    except ValueError:
        print("Error: Age must be a number!")
        return

    email = input("Enter email: ").strip()

    if not email:
        print("Error: Email cannot be empty!")
        return

    new_user = {
        "username": username,
        "password": password,
        "age": age,
        "email": email
    }

    users.append(new_user)

    save_users()

    print(f"Success! Account '{username}' registered.")
def login():

    print("\n--- Login ---")

    username = input("Enter username: ").strip()
    password = input("Enter password: ").strip()

    if not username and not password:
        print("Error: Enter username and password!")
        return

    if not username or not password:
        print("Error: Username and password required!")
        return

    for user in users:

        if user["username"] == username:

            if user["password"] == password:

                print(f"Welcome back, {username}!")
                return

            else:
                print("Error: Invalid password!")
                return

    print("Error: Username not found!")
def display_users():

    print("\n--- All Registered Users ---")

    if not users:
        print("No users registered yet.")
        return

    sorted_users = sorted(users, key=lambda u: u["username"])

    for i, user in enumerate(sorted_users, 1):

        print(
            f"{i}. Username: {user['username']} | "
            f"Age: {user['age']} | "
            f"Email: {user['email']}"
        )
def main():

    load_users()

    print("=== Login Management System ===")

    while True:

        print("\n1. Register")
        print("2. Login")
        print("3. View All Users")
        print("4. Exit")

        try:
            choice = int(input("Enter choice (1-4): "))

        except ValueError:
            print("Error: Enter number only!")
            continue

        if choice == 1:
            register()

        elif choice == 2:
            login()

        elif choice == 3:
            display_users()

        elif choice == 4:
            print("Goodbye!")
            break

        else:
            print("Error: Invalid choice!")
if __name__ == "__main__":
    main()
```
