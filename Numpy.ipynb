# Do Math Functions using NUMPY 

```python
# add tw arrays
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
print(arr1+arr2)
```

```python
# sub two arrays
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
print(arr1-arr2)
```

```python
# multiply tw arrays
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
print(arr1*arr2)
```

```python
# div tw arrays
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
print(arr1/arr2)
```

```python
# dot mutiply
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
print(arr1@arr2)
```

```python
#  max number arrays
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
print(arr1.max())
print(arr2.max())
```

```python
#  max number arrays
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([[6,7,8,9],[7,8,9,12,]])
print(arr1.max(axis=0))
print(arr2.max(axis=1))
```

```python
#  min number arrays
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([6,7,8,9])
print(arr1.min())
print(arr2.min())
```
