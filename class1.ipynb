```python
#        AI Recruitment Decision System

name = input("Enter Candidate Name: ")

age = int(input("Enter Age: "))
cgpa = float(input("Enter CGPA: "))
skills = int(input("Enter Skills Count: "))
experience = int(input("Enter Experience (Years): "))
salary = int(input("Enter Expected Salary: "))
communication = int(input("Enter Communication Score (1-10): "))

if age < 18:
    print("Rejected: Candidate must be at least 18 years old.")

elif cgpa < 2.7:
    print("Rejected: CGPA is below minimum requirement.")

elif skills < 3:
    print("Rejected: Minimum 3 skills required.")

else:

    if salary > 400000 and experience == 0:
        print("⚠ Suspicious Candidate Detected!")
        print("Reason: High salary demand with zero experience.\n")
   
    if (cgpa >= 3.3 and skills >= 7 and communication >= 7):
        level = "Software Engineer"

    elif (cgpa >= 3.0 and skills >= 5 and experience >= 1):
        level = "Junior Developer"

    elif (cgpa >= 2.7 and skills >= 3):
        level = "Intern"

    else:
        level = "Not Selected"
    
    print(f"Candidate Name : {name}")
    print(f"Hiring Status  : {level}")
```
