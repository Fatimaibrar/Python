```python
#1. Library Book Management System by using classs
class librarybook:
    def __init__(self, title, authors, availablecopies):
        self.title= title  #public
        self.author= authors   #public
        self._availablecopies= availablecopies #protected 
    def borrowBook(self):
        if self._availablecopies > 0:
            print("No of copies", {self._availablecopies-1} ) 
    def returnBook(self):            
             print("No of books", {self._availablecopies+1} )
    def showbookdetail(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
    def show(self):
        print("Welcome to the Library!")    
        print(f"Available Copies: {self._availablecopies}")
        print(f"Title: {self.title},\nAuthor: {self.author},\nAvailable Books: {self._availablecopies}")    
    
```

```python
book1 = librarybook("zawiya", "ishfaq ahmad", 5)
book1.show()
book1.borrowBook()
book1.returnBook()
print("\n")
book2 = librarybook("naml","umera ahmad",7)
book2.show()
book2.borrowBook()
book2.returnBook()
print("\n")
book3 = librarybook("yarm","nimra ahmad",2)
book3.show()
book3.borrowBook()
book3.returnBook()
```

```python
# Employee Management System
class Employee:
 
    def __init__(self, name, designation, salary):
        self.name = name                    # public
        self.designation = designation      # public
        self._salary = salary               # protected

    # Show employee details
    def showDetails(self):
        print(f"Name: {self.name}")
        print(f"Designation: {self.designation}")

    # Increase salary
    def increaseSalary(self, amount):
        self._salary += amount
        print(f"Salary increased by Rs.{amount}")
    # Show current salary
    def showSalary(self):
        print(f"Current Salary: Rs.{self._salary}")


# Create 3 employee objects
emp1 = Employee("Ali", "Manager", 50000)
emp2 = Employee("Sara", "Developer", 40000)
emp3 = Employee("Ahmed", "Designer", 35000)

# Display details
print("Employee 1:")
emp1.showDetails()
emp1.showSalary()

print("\nEmployee 2:")
emp2.showDetails()
emp2.showSalary()

print("\nEmployee 3:")
emp3.showDetails()
emp3.showSalary()

# Increase salary of one employee
print("\nIncrementing employees's Salary...")
emp2.increaseSalary(5000)

# Show updated salary
emp2.showSalary()
```
