# Student Management System

```python
students = []
used_ids = set()
```

```python
def get_grade(marks):
    if marks >= 80:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "Fail"
get_grade(85)    
```

```python
def add_student():
    print("\n--- Add Student ---")

    try:
        student_id = int(input("Enter Student ID: "))
    except ValueError:
        print("Error: Student ID must be a number!")
        return

    if student_id in used_ids:
        print("Error: Student ID already exists!")
        return

    name = input("Enter Name: ").strip()
    if not name:
        print("Error: Name cannot be empty!")
        return

    try:
        age = int(input("Enter Age: "))
    except ValueError:
        print("Error: Age must be a number!")
        return

    course = input("Enter Course: ").strip()

    try:
        marks = float(input("Enter Marks (0-100): "))
        if marks < 0 or marks > 100:
            print("Error: Marks must be between 0 and 100!")
            return
    except ValueError:
        print("Error: Marks must be a number!")
        return

    grade = get_grade(marks)

    student = {
        "id": student_id,
        "name": name,
        "age": age,
        "course": course,
        "marks": marks,
        "grade": grade
    }

    students.append(student)
    used_ids.add(student_id)
    print(f"Student '{name}' added successfully with grade {grade}.")
add_student()   
```

```python
def display_students():
    print("\n--- All Students ---")
    if not students:
        print("No students found.")
        return
    for s in students:
        print(f"ID: {s['id']} | Name: {s['name']} | Age: {s['age']} | "
              f"Course: {s['course']} | Marks: {s['marks']} | Grade: {s['grade']}")
display_students()        
```

```python
def search_student():
    print("\n--- Search Student ---")
    try:
        search_id = int(input("Enter Student ID to search: "))
    except ValueError:
        print("Error: ID must be a number!")
        return

    for s in students:
        if s["id"] == search_id:
            print(f"Found: ID: {s['id']} | Name: {s['name']} | "
                  f"Marks: {s['marks']} | Grade: {s['grade']}")
            return

    print("Student not found!")
search_student()    
```

```python
def update_marks():
    print("\n--- Update Student Marks ---")
    try:
        search_id = int(input("Enter Student ID to update: "))
    except ValueError:
        print("Error: ID must be a number!")
        return

    for s in students:
        if s["id"] == search_id:
            try:
                new_marks = float(input("Enter new marks (0-100): "))
                if new_marks < 0 or new_marks > 100:
                    print("Error: Marks must be between 0 and 100!")
                    return
            except ValueError:
                print("Error: Marks must be a number!")
                return

            s["marks"] = new_marks
            s["grade"] = get_grade(new_marks)
            print(f"Marks updated. New grade: {s['grade']}")
            return

    print("Student not found!")
update_marks()    
```

```python
def delete_student():
    print("\n--- Delete Student ---")
    try:
        search_id = int(input("Enter Student ID to delete: "))
    except ValueError:
        print("Error: ID must be a number!")
        return

    for s in students:
        if s["id"] == search_id:
            students.remove(s)
            used_ids.discard(search_id)
            print(f"Student ID {search_id} deleted.")
            return

    print("Student not found!")
delete_student()    
    
```

```python
def sort_students():
    print("\n1. Sort by Marks\n2. Sort by Name")
    choice = input("Choose: ").strip()

    if choice == "1":
        sorted_list = sorted(students, key=lambda s: s["marks"], reverse=True)
        print("\n--- Students Sorted by Marks (High to Low) ---")
    elif choice == "2":
        sorted_list = sorted(students, key=lambda s: s["name"])
        print("\n--- Students Sorted by Name (A-Z) ---")
    else:
        print("Invalid choice!")
        return

    for s in sorted_list:
        print(f"{s['name']} | Marks: {s['marks']} | Grade: {s['grade']}")
sort_students()        
```

```python
def main():
    print("=== Student Management System ===")

    while True:
        print("\n1. Add Student")
        print("2. View Students")
        print("3. Search Student")
        print("4. Update Marks")
        print("5. Delete Student")
        print("6. Sort Students")
        print("7. Exit")

        try:
            choice = int(input("Enter choice (1-7): "))
        except ValueError:
            print("Error: Enter a number between 1 and 7!")
            continue

        if choice == 1:
            add_student()
        elif choice == 2:
            display_students()
        elif choice == 3:
            search_student()
        elif choice == 4:
            update_marks()
        elif choice == 5:
            delete_student()
        elif choice == 6:
            sort_students()
        elif choice == 7:
            print("Goodbye!")
            break
        else:
            print("Error: Invalid choice!")

main()
```
