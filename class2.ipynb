```python
# 1. Create variables
university_name = "Punjab University"
total_number_of_students = 0
scholarship_budget = 500000
passing_marks = 50

# 2. Create a list of at least 8 students
student_names = ["Ali", "Sara", "Ahmed", "Fatima", "Usman", "Ayesha", "Bilal", "Hina"]

# 3. Create a tuple for departments
departments = ("CS", "SE", "AI", "DS", "IT")

# List to store all students entered during input
all_students = []
passed_students = []  # 6. Store eligible students here

print(f"\n=== {university_name} Student Entry System ===")
print("Type 'exit' as student name to stop entering data.\n")

# 4. Use a loop to take input until 'exit'
while True:
    name = input("Enter student name: ")
    
    if name.lower() == "exit":
        break  # stop the loop
    
    dept = input("Enter department (CS/SE/AI/DS/IT): ")
    marks = float(input("Enter marks obtained: "))
    attendance = float(input("Enter attendance percentage: "))
    
    total_number_of_students += 1
    
    # Store student data as a dictionary in the list
    student_data = {
        "name": name,
        "department": dept,
        "marks": marks,
        "attendance": attendance
    }
    all_students.append(student_data)
   
   
  
    # 5. Apply if conditions
    if dept not in departments:
        print("Invalid Department")
    
    elif attendance < 50 or marks < 50:
        print("Student Not eligible")
    
    elif marks >= passing_marks and attendance > 75:
        print("Eligible for Exam")
        passed_students.append(name)  # add to passed list
        
        if marks > 85 and attendance > 80:
            print("Scholarship Candidate")
    
    print("----------------\n")

# 7. After loop ends - formatted output
print("\n========== SUMMARY ==========")
print(f"Total students entered: {total_number_of_students}")

print("\nPassed/Eligible Students:")
if passed_students:
    for i, s in enumerate(passed_students, 1):
        print(f"{i}. {s}")
else:
    print("No student is eligible.")

print("\nAvailable Departments:")
for dept in departments:
    print(f"- {dept}")

# Find student with highest marks
if all_students:
    top_student = max(all_students, key=lambda x: x["marks"])
    print(f"\nStudent with Highest Marks:")
    print(f"Name: {top_student['name']}")
    print(f"Department: {top_student['department']}")
    print(f"Marks: {top_student['marks']}")
    print(f"Attendance: {top_student['attendance']}%")
else:
    print("\nNo student data entered.")

print("==============================")
```
