```python
#Create a dictionary and then have list, tuple and set as values in it. After that you have to access each element from list, set and tuple
my_dict = {
    'my_list': [1, 2, 3, 4, 5],
    'my_tuple': ('a', 'b', 'c', 'd', 'e'),
    'my_set': {10, 20, 30, 40, 50}
}
# Accessing elements from the list
print("Elements in the list:")
for element in my_dict['my_list']:
    print(element)
print("Elements in the tuple:")
for element in my_dict['my_tuple']:
    print(element)
print("Elements in the set:")
for element in my_dict['my_set']:
    print(element)
```

```python
#Create a list, in it you have a dictionary, tuple and set and access its elements
my_list = [
    {'name': 'fatima', 'age': 24},   
    ('x', 'y', 'z'),                
    {100, 200, 300}     
]
# Accessing elements from the dictionary
print("Elements in the dictionary:")
for key, value in my_list[0].items():
    print(f"{key}: {value}")

# Accessing elements from the tuple
print("Elements in the tuple:")
for element in my_list[1]:
    print(element)

# Accessing elements from the set
print("Elements in the set:")
for element in my_list[2]:
    print(element)
```

```python
#Create a Tuple, in it you have a dictionary, List and set and access its elements
my_tuple = (
    {'name': 'fatima', 'age': 24},   
    ['x', 'y', 'z'],                
    {100, 200, 300}     
)
# Accessing elements from the dictionary
print("Elements in the dictionary:")
for key, value in my_tuple[0].items():
    print(f"{key}: {value}")

# Accessing elements from the list
print("Elements in the list:")
for element in my_tuple[1]:
    print(element)

# Accessing elements from the set
print("Elements in the set:")
for element in my_tuple[2]:
    print(element)
```

```python
#Create a Set, in it you have a dictionary, tuple and list and access its elements
my_set = {
    ('x', 'y', 'z'),
    ('a', 'b', 'c')
}
# Accessing elements from the tuple
print("Elements in the tuple:")
for element in my_set:
    print(element)
```

```python
# Create a Set, in it you have a dictionary, tuple and list and access its elements

# Create a tuple (hashable)
my_tuple = (1, 2, 3, "apple")

# Create a set with the tuple
my_set = {my_tuple, "hello", 42}

print(f"Set: {my_set}")
print(f"Access tuple from set: {my_tuple}")
print(f"Access tuple elements: {my_tuple[0]}, {my_tuple[1]}, {my_tuple[2]}")

print("\n" + "=" * 50)
print("WORKAROUND: Store reference to Dictionary and List")
print("=" * 50)

# Create dictionary and list separately (not in set)
my_dict = {"name": "Ali", "age": 20}
my_list = [10, 20, 30, 40]

# Access elements
print(f"Dictionary: {my_dict}")
print(f"Access dict element: {my_dict['name']}")

print(f"\nList: {my_list}")
print(f"Access list element: {my_list[0]}, {my_list[2]}")

# Create a set with the tuple, and keep references to dict and list
my_set = {my_tuple, "data"}

print(f"\nSet with tuple: {my_set}")
print(f"Tuple in set: {my_tuple}")

try:
    # This will cause an error
    bad_set = {my_dict}
    print("Dictionary added successfully")
except TypeError as e:
    print(f"Error: {e}")
    print("Reason: Dictionary is unhashable!")

try:
    # This will also cause an error
    bad_set = {my_list}
    print("List added successfully")
except TypeError as e:
    print(f"Error: {e}")
    print("Reason: List is unhashable!")
```
